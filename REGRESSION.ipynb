```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
import joblib
```

```python
df = pd.read_csv("Ice_Cream_Temperature.csv")
```

## EDA  AND PREPROCESSING 
preprocess data but this data is preprocessed 

## REGRESSION
##### 1) split data
train_test_split(x,y,test_size=0.20,random_state=42)   
splits data int 80% training and 20% testing ratio  
random_state removes randomness

```python
x=df.drop(columns=['Ice Cream Profits'])
y=df['Ice Cream Profits']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

##### 2) create model 
##### 3) training initializer
##### 4) testing initializer

```python
knn =KNeighborsRegressor(n_neighbors=3) #create knn model
knn.fit(x_train, y_train) # trainnig
y_pred=knn.predict(x_test) # testing
```

##### 5) confusion matrix
actual value= y  
predicted =y^ 
summation (y-y^)/n   
tells average error between 2 values 

```python
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R²): {r2}")
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape}")
```
