```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import joblib
```

```python
df = pd.read_csv("Heart_Disease_Prediction_Updated.csv")
```

## EDA  AND PREPROCESSING 
preprocess data but this data is preprocessed

## CLASSIFICATION
##### 1) split data
train_test_split(x,y,test_size=0.20,random_state=42)   
splits data int 80% training and 20% testing ratio  
random_state removes randomness

```python
x=df.drop(columns=['Heart Disease'])
y=df['Heart Disease']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

##### 2) create model 
##### 3) training initializer
##### 4) testing initializer

```python
nb = GaussianNB() #create guasssian model
nb.fit(x_train, y_train) # trainnig
y_pred=nb.predict(x_test) # testing
```

##### 5) confusion matrix
###### accuracy
ratio true positive and true negative in a model TP+TN/N
###### precisiion 
ratio of true positives from predicted positives TP+/TP+FP
###### recall
ratio true positive from predicted outputs in a model TP/TP+FN
###### f1=2PR/P+R 

```python

#BINARY CLASSIFICATION
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
precision = precision_score(y_test, y_pred, average='binary') #macro for multi class
print(f"Precision: {precision}")
recall = recall_score(y_test, y_pred, average='binary')
print(f"Recall: {recall}")
f1 = f1_score(y_test, y_pred, average='binary')
print(f"F1 Score: {f1}")
conf_matrix = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(conf_matrix).plot()
```

```python
nb =KNeighborsClassifier(n_neighbors=3) #create knn model
nb.fit(x_train, y_train) # trainnig
y_pred=nb.predict(x_test) # testing
#BINARY CLASSIFICATION
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
precision = precision_score(y_test, y_pred, average='binary')
print(f"Precision: {precision}")
recall = recall_score(y_test, y_pred, average='binary')
print(f"Recall: {recall}")
f1 = f1_score(y_test, y_pred, average='binary')
print(f"F1 Score: {f1}")
conf_matrix = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(conf_matrix).plot()
```
